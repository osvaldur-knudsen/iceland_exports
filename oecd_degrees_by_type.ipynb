```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.options.display.max_rows = 999
```

```python
# Read data from csv files
df_degrees = pd.read_csv('oecd_degrees_by_type.csv')
df_degrees
```

```python
df_degrees.loc[18, 'Information and Communication Technologies'] = 0
df_degrees['Information and Communication Technologies'] = df_degrees['Information and Communication Technologies'].astype('float')
```

```python
filter = df_degrees['Country']=='Iceland'
df_degrees[filter]
```

```python
df_degrees['Science degrees'] = (df_degrees['Natural sciences, mathematics and statistics']+
                                    df_degrees['Information and Communication Technologies']+
                                    df_degrees['Engineering, manufacturing and construction'])
```

```python
df_degrees.head()
```

```python
df_oecd = df_degrees.describe()
```

```python
df_oecd = df_oecd.loc['mean',:]
```

```python
df_oecd.values
```

```python
df_degrees = df_degrees.append(df_oecd)
```

```python
df_degrees = df_degrees.reset_index(drop = True)
```

```python
df_degrees.loc[37,'Country'] = 'OECD'
```

```python
df_degrees = df_degrees[['Country','Science degrees']]
df_degrees
```

```python
df_degrees.to_excel('df_oecd_science_degrees.xlsx')
```
